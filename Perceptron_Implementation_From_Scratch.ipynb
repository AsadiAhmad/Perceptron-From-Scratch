# Step7: Implement Perceptron

```python
def sign(z):
    if z < 0:
        return -1
    else:
        return 1
```

```python
def update_weights(B, W1, W2, X1, X2, Y):
    w1 = W1 + X1*Y
    w2 = W2 + X2*Y
    b = B + Y
    return b, w1, w2
```

```python
def perceptron(train_dataset, test_dataset, epochs):
    err = 0
    W1 = 0
    W2 = 0
    B = 0
    last_W1 = -1
    last_W2 = -1
    last_B = -1
    for epoch in range(epochs):
        err = 0
        for row in range(len(train_dataset)):
            X1 = train_dataset['feature1'][row]
            X2 = train_dataset['feature2'][row]
            Y = train_dataset['label'][row]
            if sign(X1 * W1 + X2 * W2 + B) != Y:
                B, W1, W2 = update_weights(B, W1, W2, X1, X2, Y)
                err += 1
        print(f"epoch: {epoch + 1}, Accuracy: {(len(train_dataset) - err)/ len(train_dataset)}")
        print("----------------------------------")
        if (last_W1 == W1 and last_W2 == W2 and last_B == B and err == 0):
            return W1, W2, B, epoch + 1
        else:
             last_W1 = W1
             last_W2 = W2
             last_B = B
    return W1, W2, B, epoch + 1
```

```python
weight1_1, weight1_2, bayas1, epoch1 = perceptron(train_dataset1, test_dataset1, 20)
```

```python
weight2_1, weight2_2, bayas2, epoch2 = perceptron(train_dataset2, test_dataset2, 20)
```

# Step8: Calclate the Confusion Matrix

```python
def calculate_confusion_matrix(dataset, weight1, weight2, bayas):
    if 'predicted_label' not in dataset.columns:
        dataset['predicted_label'] = None
    if 'result' not in dataset.columns:
        dataset['result'] = None

    TP = 0
    FN = 0
    FP = 0
    TN = 0

    for row in range(len(dataset)):
        dataset.loc[row, 'predicted_label'] = sign((dataset.loc[row, 'feature1'] * weight1) + (dataset.loc[row, 'feature2'] * weight2) + bayas)
        Label = dataset.loc[row, 'label']
        Predicted_label = dataset.loc[row, 'predicted_label']
        if Label == 1 and Predicted_label == 1:
            dataset.loc[row, 'result'] = 'TP'
            TP += 1
        elif Label == 1 and Predicted_label == -1:
            dataset.loc[row, 'result'] = 'FN'
            FN += 1
        elif Label == -1 and Predicted_label == 1:
            dataset.loc[row, 'result'] = 'FP'
            FP += 1
        elif Label == -1 and Predicted_label == -1:
            dataset.loc[row, 'result'] = 'TN'
            TN += 1

    dataset = dataset[['feature1', 'feature2', 'label', 'predicted_label', 'result']]
    return TP, FN, FP, TN, dataset
```

```python
TP_Tr1, FN_Tr1, FP_Tr1, TN_Tr1, train_dataset1 = calculate_confusion_matrix(train_dataset1, weight1_1, weight1_2, bayas1)
TP_Tr2, FN_Tr2, FP_Tr2, TN_Tr2, train_dataset2 = calculate_confusion_matrix(train_dataset2, weight2_1, weight2_2, bayas2)
TP1, FN1, FP1, TN1, test_dataset1 = calculate_confusion_matrix(test_dataset1, weight1_1, weight1_2, bayas1)
TP2, FN2, FP2, TN2, test_dataset2 = calculate_confusion_matrix(test_dataset2, weight2_1, weight2_2, bayas2)
```

```python
def show_confusion_matrix(tp, fn, fp, tn):
    conf_matrix = np.array([[tp, fn], [fp, tn]])
    fig, ax = plt.subplots()

    cax = ax.matshow(conf_matrix, cmap='Blues')

    plt.colorbar(cax)

    for (i, j), val in np.ndenumerate(conf_matrix):
        ax.text(j, i, f'{val}', ha='center', va='center', color='black')

    ax.set_xticks([0, 1])
    ax.set_yticks([0, 1])

    ax.set_xticklabels(['Predicted Positive', 'Predicted Negative'])
    ax.set_yticklabels(['Actual Positive', 'Actual Negative'])

    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title('2x2 Confusion Matrix')

    plt.show()
```

# Step9: Calculate Measurements

```python
def calculate_measurements(tp, fn, fp, tn):
    precision = tp / (tp + fp) if tp + fp != 0 else 0
    recall = tp / (tp + fn) if tp + fn != 0 else 0
    accuracy = (tp + tn) / (tp + tn + fp + fn) if tp + tn + fp + fn != 0 else 0
    f_measure = 2 * (precision * recall) / (precision + recall) if precision + recall != 0 else 0

    precision *= 100
    recall *= 100
    accuracy *= 100
    f_measure *= 100

    print(f"Precision: {precision:.2f}%")
    print(f"Recall: {recall:.2f}%")
    print(f"Accuracy: {accuracy:.2f}%")
    print(f"F-Measure (F1 Score): {f_measure:.2f}%")
    return precision, recall, accuracy, f_measure
```

```python
calculate_measurements(TP1, FN1, FP1, TN1)
```

```python
calculate_measurements(TP_Tr2, FN_Tr2, FP_Tr2, TN_Tr2)
```
